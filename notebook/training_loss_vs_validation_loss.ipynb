```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def generate_loss_plot(s, version):
    loss = []
    loss_val = []

    for i, e in enumerate(s.split('\n')):
        if (i-2)%3 == 0:
            loss.append(float(e[e.find('loss='):e.find('\t v')].strip('loss=')))
            loss_val.append(float(e[e.find('val_loss='):e.find(' \t t')].strip('val_loss=')))
    
    l = len(loss)
    x_epoch = range(1, l + 1)

    # plot train and validation loss
    plt.plot(x_epoch, loss)
    plt.plot(x_epoch, loss_val)
    plt.title('model train vs validation loss - version{}'.format(version))
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.xlim((0, l+1)) 
    plt.ylim((0.3, 0.4)) 
    plt.legend(['train', 'validation'], loc='upper right')
    plt.grid()
    plt.show()
```
